```python
import pandas as pd
import numpy
import math
from scipy import stats
```

```python
def two_samp_proportion(p1,p2,n1,n2):
    p_pool = ((p1*n1)+(p2*n2))/(n1+n2)
    s_sq = (p_pool*(1-p_pool)*((1/n1)+(1/n2)))
    s = math.sqrt(s_sq)
    z = (p1-p2)/s
    if(z<0):
        p_val = stats.norm.cdf(z)
    else:
        p_val = 1-stats.norm.cdf(z)
        return z, p_val*2
```

```python
two_samp_proportion(0.27,0.19,100,100)
```
